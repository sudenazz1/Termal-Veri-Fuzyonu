# 01 — Data Exploration
**TÜBİTAK 2209-A | Thermal Bridge Detection**

This notebook explores the TBBR dataset structure, statistics, and visualizes thermal images with annotations.

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

BASE = '/content/drive/MyDrive/bitirme_projesi'

with open(f'{BASE}/Flug1_100-104Media_coco.json') as f:
    train_json = json.load(f)
with open(f'{BASE}/Flug1_105Media_coco.json') as f:
    test_json = json.load(f)

print('Keys:', list(train_json.keys()))
```

```python
print('TRAIN')
print(f'  Images      : {len(train_json["images"])}')
print(f'  Annotations : {len(train_json["annotations"])}')
print(f'  Classes     : {[c["name"] for c in train_json["categories"]]}')
print()
print('TEST')
print(f'  Images      : {len(test_json["images"])}')
print(f'  Annotations : {len(test_json["annotations"])}')
```

```python
img_ids = [ann['image_id'] for ann in train_json['annotations']]
ann_per_image = Counter(img_ids)
valeurs = list(ann_per_image.values())

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].hist(valeurs, bins=20, color='steelblue', edgecolor='white')
axes[0].set_title('Annotations per image')
axes[0].set_xlabel('Number of thermal bridges')

aires = [ann['area'] for ann in train_json['annotations']]
axes[1].hist(aires, bins=40, color='coral', edgecolor='white')
axes[1].set_title('Area distribution')
axes[1].set_xlabel('Area (px²)')

largeurs = [ann['bbox'][2] for ann in train_json['annotations']]
hauteurs  = [ann['bbox'][3] for ann in train_json['annotations']]
axes[2].scatter(largeurs, hauteurs, alpha=0.3, s=10, color='teal')
axes[2].set_title('Width vs Height')
plt.tight_layout()
plt.show()
```
