```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models

# --- CONNEXION DRIVE ---
from google.colab import drive
drive.mount('/content/drive')

# --- CONFIGURATION DES CHEMINS ---
# Vérifiez bien que le nom du dossier sur votre Drive est 'Projet_Tubitak'
PATH_IMAGES = "/content/drive/MyDrive/tubitak/imaj"
PATH_MASKS = "/content/drive/MyDrive/tubitak/masks"

IMG_HEIGHT = 128
IMG_WIDTH = 128
```

```python
def load_data():
    X, Y = [], []
    mask_files = [f for f in os.listdir(PATH_MASKS) if f.endswith('.png')]
    print(f"Chargement de {len(mask_files)} images depuis le Drive...")

    for mask_name in mask_files:
        try:
            # 1. Charger le Masque (LabelMe)
            mask_path = os.path.join(PATH_MASKS, mask_name)
            mask = cv2.imread(mask_path, 0)
            mask = cv2.resize(mask, (IMG_HEIGHT, IMG_WIDTH))
            mask = np.expand_dims(mask, axis=-1) / 255.0
            mask[mask > 0.5] = 1
            mask[mask <= 0.5] = 0

            # 2. Chercher l'image thermique correspondante
            base_name = os.path.splitext(mask_name)[0]
            img_path = os.path.join(PATH_IMAGES, base_name + ".jpg")
            if not os.path.exists(img_path):
                img_path = os.path.join(PATH_IMAGES, base_name + ".JPG")

            img = cv2.imread(img_path)
            img = cv2.resize(img, (IMG_HEIGHT, IMG_WIDTH))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            X.append(img)
            Y.append(mask)
        except Exception as e:
            continue

    return np.array(X), np.array(Y)

X, Y = load_data()
# Normalisation et séparation (90% entraînement, 10% test)
X_train, X_test, Y_train, Y_test = train_test_split(X / 255.0, Y, test_size=0.1, random_state=42)
print(f"Prêt ! Entraînement sur {len(X_train)} images.")
```

```python
def build_unet():
    inputs = layers.Input((IMG_HEIGHT, IMG_WIDTH, 3))

    # ENCODEUR (Analyse l'image)
    c1 = layers.Conv2D(32, (3,3), activation='relu', padding='same')(inputs)
    p1 = layers.MaxPooling2D((2,2))(c1)

    # BOTTLENECK
    c2 = layers.Conv2D(64, (3,3), activation='relu', padding='same')(p1)

    # DÉCODEUR (Reconstruit le masque)
    u3 = layers.Conv2DTranspose(32, (2,2), strides=(2,2), padding='same')(c2)
    outputs = layers.Conv2D(1, (1,1), activation='sigmoid')(u3)

    model = models.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Lancement de l'entraînement
model = build_unet()
# 50 époques suffisent pour 87 images sur GPU
history = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=50, batch_size=8)

# Sauvegarde du cerveau de l'IA sur votre Drive
model.save('/content/drive/MyDrive/tubitak/model_energie_israfi.h5')
```

```python
# --- VISUALISATION DE L'ENTRAÎNEMENT ---
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(1, len(acc) + 1)

    plt.figure(figsize=(15, 5))

    # Courbe d'Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Entraînement Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.title('Précision (Accuracy) par Époque')
    plt.xlabel('Époques')
    plt.ylabel('Précision')
    plt.legend(loc='lower right')

    # Courbe de Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Entraînement Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.title('Perte (Loss) par Époque')
    plt.xlabel('Époques')
    plt.ylabel('Perte')
    plt.legend(loc='upper right')

    plt.show()

plot_training_history(history)
```

```python
from tensorflow.keras.metrics import MeanIoU

def visualize_results_with_iou(X_test, Y_test, model, num_samples=5):
    # Calcul des prédictions
    preds_test = model.predict(X_test)
    preds_test_t = (preds_test > 0.5).astype(np.uint8) # Seuil à 0.5

    for i in range(num_samples):
        # Calcul de l'IoU pour cette image précise
        intersection = np.logical_and(Y_test[i], preds_test_t[i])
        union = np.logical_or(Y_test[i], preds_test_t[i])

        # Éviter la division par zéro si l'image est toute noire
        if np.sum(union) == 0:
            iou_score = 1.0
        else:
            iou_score = np.sum(intersection) / np.sum(union)

        # Affichage
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.imshow(X_test[i])
        plt.title(f"Image Thermique #{i+1}")
        plt.axis('off')

        plt.subplot(1, 3, 2)
        plt.imshow(Y_test[i].squeeze(), cmap='gray')
        plt.title("Masque Réel (LabelMe)")
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.imshow(preds_test_t[i].squeeze(), cmap='gray')
        plt.title(f"Prédiction IA (IoU: {iou_score:.4f})")
        plt.axis('off')

        plt.show()

# Appeler la fonction pour voir les 5 premières images de test
visualize_results_with_iou(X_test, Y_test, model, num_samples=5)
```

```python
import pandas as pd

def quantify_results(X_test, Y_test, model):
    preds = (model.predict(X_test) > 0.5).astype(np.uint8)
    results = []

    for i in range(len(preds)):
        # Calcul de la surface de perte (pixels blancs)
        perte_pixels = np.sum(preds[i] == 1)
        total_pixels = preds[i].size
        ratio_perte = (perte_pixels / total_pixels) * 100

        results.append({
            "Image_ID": i,
            "Surface_Perte_Pixels": perte_pixels,
            "Ratio_Perte_%": round(ratio_perte, 2)
        })

    df = pd.DataFrame(results)
    print("--- QUANTIFICATION DES PERTES THERMIQUES ---")
    return df

df_pertes = quantify_results(X_test, Y_test, model)
print(df_pertes)
```

```python
from tensorflow.keras import backend as K

# Calcul du Dice Coefficient (mentionné p.5 du rapport)
def dice_coef(y_true, y_pred, smooth=1):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

# Évaluation sur toutes les images de test
print("--- BILAN SCIENTIFIQUE FINAL (87 IMAGES) ---")
results = model.evaluate(X_test, Y_test, verbose=0)
preds_test = (model.predict(X_test) > 0.5).astype(np.float32)

# Calcul manuel du Dice moyen
dice_scores = [dice_coef(Y_test[i].astype(np.float32), preds_test[i]).numpy() for i in range(len(X_test))]

print(f"Précision Globale (Accuracy) : {results[1]*100:.2f}%")
print(f"Dice Coefficient Moyen (Cible TÜBİTAK) : {np.mean(dice_scores):.4f}")
# Note: L'IoU est mathématiquement lié au Dice.
print(f"Mean IoU estimé : {np.mean(dice_scores)/(2-np.mean(dice_scores)):.4f}")
```

```python
import pandas as pd

# --- CONFIGURATION PHYSIQUE (Basée sur Zenodo & Littérature) ---
TEMP_EXT = 4.37  # Moyenne du dataset Karlsruhe
TEMP_INT = 20.0  # Hypothèse de confort
DELTA_T = TEMP_INT - TEMP_EXT

# Coût moyen de l'énergie en Turquie (Estimation pour le projet)
PRIX_KWH_TL = 2.50  # À ajuster selon les tarifs actuels

def estimer_gaspillage(df_ia):
    """
    Simule la fusion de données (Faaliyet 4)
    Transforme le ratio de pixels blancs en perte énergétique estimée.
    """
    # Hypothèse : Un toit de 100m2 représenté par les 640x512 pixels
    # On calcule la perte basée sur un coefficient de transmission thermique (U)
    # pour un pont thermique (ex: 2.0 W/m2.K)

    resultats_fusion = []

    for index, row in df_ia.iterrows():
        ratio = row['Surface_Perte_Relative_%'] # Changed 'Zone_de_Perte_%' to 'Surface_Perte_Relative_%'

        # 1. Estimation de la puissance perdue (Watts)
        # Formule simplifiée : Surface * U * DeltaT
        # Ici on utilise le ratio pour pondérer une perte maximale théorique
        puissance_perdue_w = ratio * 15.0 * DELTA_T  # 15.0 est un facteur d'échelle

        # 2. Conversion en kWh (sur un mois de chauffage)
        conso_mensuelle_kwh = (puissance_perdue_w * 24 * 30) / 1000

        # 3. Conversion en TL
        cout_mensuel_tl = conso_mensuelle_kwh * PRIX_KWH_TL

        resultats_fusion.append({
            "Image_ID": row['Image_Index'], # Changed 'Image_ID' to 'Image_Index'
            "Ratio_Fuite_%": ratio,
            "Perte_Estimee_kWh_Mois": round(conso_mensuelle_kwh, 2),
            "Surcout_Estime_TL_Mois": round(cout_mensuel_tl, 2)
        })

    return pd.DataFrame(resultats_fusion)

# Application sur vos résultats d'IA
df_final_fusion = estimer_gaspillage(df_final) # Changed 'df_resultats' to 'df_final'
print("--- RÉSULTATS DE LA FUSION (FAALİYET 4) ---")
print(df_final_fusion)
```

```python
import cv2
import numpy as np
import os

PATH_IMAGES = "/content/drive/MyDrive/tubitak/imaj"
SAVE_DIR = "/content/drive/MyDrive/tubitak/prelabeled_output"
SAVE_ORIG_DIR = os.path.join(SAVE_DIR, "original_images")
SAVE_MASKS_DIR = os.path.join(SAVE_DIR, "generated_masks")

# Ensure output directories exist
os.makedirs(SAVE_ORIG_DIR, exist_ok=True)
os.makedirs(SAVE_MASKS_DIR, exist_ok=True)

def pre_labeling_thermique(image_path, sensibilite=210):
    img = cv2.imread(image_path, 0)

    if img is None or img.size == 0:
        return None, None

    blur = cv2.GaussianBlur(img, (5, 5), 0)
    _, mask = cv2.threshold(blur, sensibilite, 255, cv2.THRESH_BINARY)
    kernel = np.ones((3,3), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    return img, mask

print(f"Processing and saving images from: {PATH_IMAGES}")

if os.path.exists(PATH_IMAGES):
    image_files = [f for f in os.listdir(PATH_IMAGES) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    print(f"Found {len(image_files)} image files. Saving original images and generated masks...")

    processed_count = 0
    for image_filename in image_files:
        full_image_path = os.path.join(PATH_IMAGES, image_filename)
        img_orig, masque_detecte = pre_labeling_thermique(full_image_path, sensibilite=225) # Using 225 as default sensibilite

        if img_orig is not None and masque_detecte is not None:
            # Save original image
            save_orig_path = os.path.join(SAVE_ORIG_DIR, image_filename)
            cv2.imwrite(save_orig_path, img_orig)

            # Save generated mask (as grayscale PNG)
            mask_filename = os.path.splitext(image_filename)[0] + "_mask.png"
            save_mask_path = os.path.join(SAVE_MASKS_DIR, mask_filename)
            cv2.imwrite(save_mask_path, masque_detecte)

            processed_count += 1
        else:
            print(f"Could not process or save {image_filename}")

    print(f"Finished processing and saving {processed_count} images and their masks to {SAVE_DIR}")
else:
    print("Error: PATH_IMAGES does not exist. Please check your Google Drive path and ensure it's mounted.")
```

```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Configuration des chemins ---
SAVE_MASKS_DIR = "/content/drive/MyDrive/tubitak/prelabeled_output/generated_masks"

# --- CONFIGURATION PHYSIQUE (Basée sur Zenodo & Littérature) ---
TEMP_EXT = 4.37  # Moyenne du dataset Karlsruhe
TEMP_INT = 20.0  # Hypothèse de confort
DELTA_T = TEMP_INT - TEMP_EXT
PRIX_KWH_TL = 2.50  # À ajuster selon les tarifs actuels

# --- 1. Fonction pour quantifier les masques pré-labelisés ---
def quantify_prelabeled_masks(masks_dir):
    data_list = []
    mask_files = [f for f in os.listdir(masks_dir) if f.lower().endswith(('.png'))]
    mask_files.sort() # Ensure consistent ordering

    for i, mask_filename in enumerate(mask_files):
        mask_path = os.path.join(masks_dir, mask_filename)
        mask = cv2.imread(mask_path, 0) # Load as grayscale

        if mask is not None and mask.size > 0:
            perte_pixels = np.sum(mask == 255) # Count white pixels (255)
            total_pixels = mask.size
            ratio = (perte_pixels / total_pixels) * 100
            data_list.append({
                "Image_Filename": mask_filename,
                "Surface_Perte_Relative_%": round(ratio, 4),
                "Gravite": "Haute" if ratio > 10 else "Modérée" if ratio > 3 else "Faible"
            })
        else:
            print(f"Warning: Could not load or process mask file: {mask_filename}")

    return pd.DataFrame(data_list)

# --- 2. Fonction d'estimation de gaspillage (réutilisée) ---
def estimer_gaspillage(df_quantified_data):
    resultats_fusion = []
    for index, row in df_quantified_data.iterrows():
        ratio = row['Surface_Perte_Relative_%']
        puissance_perdue_w = ratio * 15.0 * DELTA_T
        conso_mensuelle_kwh = (puissance_perdue_w * 24 * 30) / 1000
        cout_mensuel_tl = conso_mensuelle_kwh * PRIX_KWH_TL

        resultats_fusion.append({
            "Image_ID": index, # Using DataFrame index as Image_ID for this context
            "Ratio_Fuite_%": ratio,
            "Perte_Estimee_kWh_Mois": round(conso_mensuelle_kwh, 2),
            "Surcout_Estime_TL_Mois": round(cout_mensuel_tl, 2)
        })
    return pd.DataFrame(resultats_fusion)

# --- 3. Fonction de tableau de bord (réutilisée) ---
def generer_dashboard_decision(df):
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(df['Surcout_Estime_TL_Mois'], bins=15, color='orange', edgecolor='black')
    plt.title('Distribution du Gaspillage Mensuel (TL) (Nouveaux Masques)')
    plt.xlabel('Surcoût estimé (TL)')
    plt.ylabel('Nombre de bâtiments/images')

    plt.subplot(1, 2, 2)
    plt.scatter(df['Ratio_Fuite_%'], df['Surcout_Estime_TL_Mois'], alpha=0.5, color='red')
    plt.title('Impact de la Surface de Fuite sur le Coût (Nouveaux Masques)')
    plt.xlabel('Ratio de Fuite (%)')
    plt.ylabel('Surcoût (TL)')

    plt.tight_layout()
    plt.savefig('dashboard_decision_prelabel_manuel.png')
    plt.show()

# --- 4. Fonction de système de décision automatique (réutilisée) ---
def systeme_decision_auto(df):
    df_copy = df.copy() # Work on a copy to avoid SettingWithCopyWarning
    df_copy['Recommandation'] = "Maintenance Mineure"
    df_copy.loc[df_copy['Surcout_Estime_TL_Mois'] > 41800, 'Recommandation'] = "URGENT : Isolation Complète Toiture"
    df_copy.loc[(df_copy['Surcout_Estime_TL_Mois'] <= 41800) & (df_copy['Surcout_Estime_TL_Mois'] > 41600), 'Recommandation'] = "Action Prioritaire : Étanchéité"
    return df_copy[['Image_ID', 'Surcout_Estime_TL_Mois', 'Recommandation']]

# --- Pipeline d'analyse pour les 352 nouveaux masques ---
print("--- Démarrage de l'analyse avec les 352 masques pré-labelisés ---")

# 1. Quantifier les masques générés
df_quantified_new_masks = quantify_prelabeled_masks(SAVE_MASKS_DIR)
print("Quantification des pertes thermiques via pré-labeling manuel (Nouveaux Masques):")
display(df_quantified_new_masks.head())

# 2. Estimer le gaspillage énergétique
df_fusion_new_masks = estimer_gaspillage(df_quantified_new_masks)
print("\n--- RÉSULTATS DE LA FUSION (Nouveaux Masques) ---")
display(df_fusion_new_masks.head())

# 3. Générer le tableau de bord
print("\nGénération du tableau de bord...")
generer_dashboard_decision(df_fusion_new_masks)

# 4. Générer le tableau de décision
df_decision_new_masks = systeme_decision_auto(df_fusion_new_masks)
print("\n--- TABLEAU DE BORD DE DÉCISION (Nouveaux Masques) ---")
display(df_decision_new_masks.head(10))

print("\nAnalyse complète effectuée pour les 352 masques.")
```
